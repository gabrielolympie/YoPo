## Component implementation

```python
yopo_prompt="""<?xml version="1.0"?>
<prompt idx="root" description="Final Implementation" predecessors="draft_1" content="Use the feedback and the First Snippet to implement a new version of the component, in a single jsx file. Return only the snippet, nothing else.">
    <prompt idx="feedback" description="Feedback" content="Using the Evaluation Criterion, make a feedback over the implemented component">
        <prompt idx="draft_1" description="First Draft" content="Implement the described component using Chakra UI and the hierarchy implemented. Put everything in a single code snippet">
            <prompt idx="hierarchy" description="Component Hierarchy" content="Propose a component hierarchy that leverage the atomic component principle. Format it as a treelib hierarchy tree.">
                <prompt idx="atomic" description="Define Atomic" content="Make a brief description of what the atomic design principle is"/>
                <prompt idx="hierarchy_draft" description="Hierarchy Draft" content="Propose a component hierarchy for the described component"/>
            </prompt>
        </prompt>
        <prompt idx="criterion" description="Evaluation Criterion" content="Based on the best practice guides, implement a checklist to evaluate a piece of chakra ui code.">
            <prompt idx="theme" description="Theme Best Practices" content="Write chakra ui best practices related to chakra ui theming" />
            <prompt idx="responsiveness" description="Responsiveness Best Practices" content="Write chakra ui best practices related to chakra ui responsiveness" />
            <prompt idx="quality" description="Code Best Practices" content="Write chakra ui best practices related to chakra ui code quality and structuration" />
            <prompt idx="hooks" description="Hooks Best Practices" content="Write chakra ui best practices related to chakra ui hooks" />
        </prompt>
    </prompt>
</prompt>"""

G=YOPO().from_xml_prompt(yopo_prompt)
G.show(figsize=(16,8))

messages=[
    {"role":"system", "content":"You are a helpful assistant"},
    {"role":"user","content":"Build a google landing page copycat in chakra ui"}
]

result = G.run(messages)

for token in result:
    print(token, end="")
```

## Strawberry

```python
yopo_prompt="""<?xml version="1.0"?>
<prompt idx="root" predecessors="method1,method2,method3,method4,method5" description="Final Answer" content="What is your final answer to my initial question?">
    <prompt idx="juge" description="Judge methods" content="Review every method we went through, and rank them from best reasoning quality to worst reasoning quality">
        <prompt idx="method1" predecessors="plan" description="Method 1" content="Proceed with the first method proposed" />
        <prompt idx="method2" predecessors="plan" description="Method 2" content="Proceed with the second method proposed" />
        <prompt idx="method3" predecessors="plan" description="Method 3" content="Proceed with the third method proposed" />
        <prompt idx="method4" predecessors="plan" description="Method 4" content="Proceed with the third method proposed" />
        <prompt idx="method5" predecessors="plan" description="Method 5" content="Proceed with the third method proposed" />
        <prompt idx="plan" description="Propose Methods" content="Propose five method to answer the question. Each method should be a perspective on the question and can be tackled just by thinking (no code or tools)" />
    </prompt>
</prompt>"""

G=YOPO().from_xml_prompt(yopo_prompt)
# compilation=G.compile()
G.show(figsize=(16,8))
```

```python
messages=[
    {"role":"system","content":"You are a helpful assistant"},
    {"role":"user","content":"How many letter 'r' are in the word 'strawberry'"},
]

result = G.run(messages)

for token in result:
    print(token, end="")
```
